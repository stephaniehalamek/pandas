## Text styling

| Property                      | Possible Values                                                           |
|-------------------------------|---------------------------------------------------------------------------|
| `fontsize` or `size`          | integer in "points" where 1 point is defaulted to 1/72nd of an inch       |
| `fontname` or `name`          | name of font as a string                                                  |
| `fontweight` or `weight`      | `'normal'`, `'bold'`, `'heavy'`, `'light'`, `'ultrabold'`, `'ultralight'` |
| `fontstyle` or `style`      | `'normal'`, `'italic'`, `'oblique'`|
| `color` or `c`                | text color                                                                |
| `backgroundcolor`             | color of rectangular background of text                                   |
| `horizontalalignment` or `ha` | `'left'`, `'center'`, `'right'`                              |
| `verticalalignment` or `va`   | `'center'`, `'top'`, `'bottom'`, `'baseline'`, `'center_baseline'`                             |
| `rotation`                    | degree of rotation                                                        |

[See all options in the documentation][1]

## Line styling

| Property                      | Possible Values                                                           |
|-------------------------------|---------------------------------------------------------------------------|
| `linewidth` or `lw`         | width of line in points       |
| `linestyle` or `ls`          | `'solid'` or `'-'` (default), `'dashed'` or `'--'`, `'dotted'` or `':'`, `'dashdot'` or `'-.'`                                                 |
| `color` or `c` | line color |
| `alpha`         | 0 to 1 - opacity       |
| `marker`        | See marker styles section       |
| `markeredgecolor` or `mec`    | marker edge color       |
| `markeredgewidth` or `mew`    | width of marker in points     |
|  `markerfacecolor` or `mfc`   | marker face color      |
|  `markersize` or `ms`   |  marker size in points      |

## Colors

Two main ways to choose colors:

* Web colors - 140 named colors known to all modern web browsers
* RGB triples - Red, Green, and Blue intensities provided as floats or hexadecimal strings

### Web colors

![0]

### RGB Triples

Red, Green, Blue intensities - 256 possible values for each. Defines many more possible colors.

* RGB floats - three-item tuple of floats between 0 and 1 - Example - `(.3, .2, .9)`
* RGB strings - two-character hexadecimal notation for each intensity as a string. Example - `#ff4c39` 

### Grayscale

* Use a string containing a float between 0 and 1. Example - `'.3'`
* Goes from black, `'0'`, to white, `'1'`.

### Color transparency

* `alpha` - float between 0 and 1 - ranges from transparent to opaque
    * Can be provided as RGBA tuple or RGBA hexadecimal string

## Matplotlib Resolution

* **Figure-Inches** - Width x Height of figure set with `figsize` - default is 6 x 4
* **Screen-Inches** - Actual width and height inches on your personal screen
* **Figure DPI** - Pixels per inch of figure - default is 72
* **Screen DPI** - Pixels per inch of your personal screen
* **Screen pixels** - Multiply figure DPI by figure inches - default is 6 * 72 by 4 * 72 or 432 x 288

### Calculating screen inches

Divide screen pixels by screen DPI. If screen DPI is 144 and you created a default figure with size 6 x 4 with 72 DPI or 432 x 288 pixels then screen inches are 432 / 144 by 288 / 144 or 3 x 2

## Text and line "points"

Matplotlib **points** (technically **typographical points**) are a measurement of size that is 1 / 72 of a figure-inch. Used for line width, edge width, marker size, font size, etc... It is independent of DPI

## Using `tick_params` to change tick labels and lines

The `tick_params` method is versatile and can change properties of both tick lines and labels

* `axis` - Choose which lines/labels to change - `'x'`, `'y'`, or `'both'`

### Parameters that change the label

* `labelsize`, `labelrotation`, `labelcolor` - change label size, degree of rotation, and color
* `labelleft`, `labelright`, `labelbottom`, `labeltop` - boolean that determines whether labels are visible are not

### Parameters that change the line

* `length` - length of tick line in points
* `width` - width of tick line in points
* `pad` - distance in points of tick label from tick line
* `direction` - 'in', 'out', or 'inout'
* `top`, `bottom`, `left`, `right` - boolean that determines whether ticks are visible.

````python
ax.tick_params(axis='y', labelright=True, labelsize=10, 
               labelrotation=15, labelcolor='red',
               length=10, width=3, pad=8, right=True)
fig
````

## Patches

Must import the `patches` module and create with constructor:

* `Circle`
* `Ellipse`
* `Rectangle`
* `Polygon`
* `Arc`

By default all patches are filled except `Arc`. Set `fill` to `False` to not fill. Here are the most common parameters:

* `fill`
* `angle` - degrees to rotate counter-clockwise
* `facecolor`/`fc`
* `edgecolor`/`ec`
* `linestyle`/`ls`
* `linewidth`/`lw`
* `hatch`

### Add patch to axes

* Use `add_patch` to add to axes
* Use `set_aspect('equal')` to make x-axis and y-axis

````python
from matplotlib import patches
fig, ax = plt.subplots(facecolor='tan')
ax.set_aspect('equal')
circle = patches.Circle((.3, .7), radius=.2, fill=False, ec='orange', lw=4, ls='--')
rectangle = patches.Rectangle((.2, .1), width=.5, height=.2, angle=20, hatch='++',
                              ec='white', fc='deeppink')
ax.add_patch(circle)
ax.add_patch(rectangle);
````

## Zorder

* Controls which plotting objects are visible when occupying the same space
* It is set to a number. Objects with a higher number are placed on top and are visible
* If objects have the same zorder, the plotting object added last is visible

### Default zorder

* patches - 1
* lines - 2
* text - 3

### Setting color cycle for axes

````python
from matplotlib import cm
fig, ax = plt.subplots()
ax.set_prop_cycle(color=cm.Dark2(range(5)))
x = np.arange(6)
y = x ** 2
ax.plot(x, y)
ax.plot(x, y + 3)
ax.plot(x, y + 6);
````
